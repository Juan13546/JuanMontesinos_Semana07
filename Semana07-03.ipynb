```python
#Utilizando el .format()
nombre = "Juan Montesinos"
edad = 19
print("Hola {} ya se que tienes {} años".format(nombre,edad))
```

```python
#Dar formato a números con muchos decimales
resultado  = 10/3
print("10/3 es = {} ".format(resultado))
print("10/3 es = {r:1.3f}".format(r=resultado))
#r:1.3f -> el r es la variable donde se almacena el valor, 1 es la cantidad de numeros enteros y el 3 la cantidad de decimales
```

```python
#Dar formato con f
#La funcion f esta dispobible desde la version 3.6
Nombre = "Juan Montesinos"
edad = 19
print(f"Buenos dias {nombre}, ya se que tienes {edad} años")
```

```python
#Dar formato a numeros con muchos decimales con f
resultado = 10/3
print(f"10/3 es = {resultado}")
print(f"10/3 es = {resultado:1.3f}")
```
